```python
import cv2
import matplotlib.pyplot as plt

imagePath = '001img.jpg'
img = cv2.imread(imagePath)

# opencv usa i colori BRG invece matplolib usa RGB, devo quindi cambiarlo solo per la preview
preview_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# uso matplotlib per la preview perch opencv  rotto 
plt.figure(figsize=(5,3))
plt.imshow(preview_img)
plt.axis('off')
```

```python
# trasformiamo l'immagine in bianco e nero
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


preview_gray_image = cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,3))
plt.imshow(preview_gray_image)
plt.axis('off')
```

```python
# carichiamo il classificatore pre trainato Haar Cascade
# messo a disposizione da opencv

# ATTENZIONE carichiamo il file .xml che immagino siano i settaggi
# per il classificatore che  trainato per riconoscere bene visi frontali
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
# ora applichiamo il classificatore alla nostra immagine

face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
```

```python
# il lavoro  fatto, per mostrare meglio il risultato disegnamo un bounding box per mostrare
# la porzione di immagine riconosciuta
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

```python
# riportiamo l'immagine a colori
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# importiamo matplotlib per mostrare l'immagine, potremmo
# farlo anche con opencv a fa cagare
import matplotlib.pyplot as plt



plt.figure(figsize=(6,3))
preview_img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)

plt.imshow(preview_img_rgb)
plt.axis('off')
```
